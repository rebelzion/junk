```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
X = torch.arange(-10, 10, 0.1)
y = X + torch.randn(X.shape)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(X, y, 'o')
```

```python
def sgd(params, lr, X, Y, epochs):
    losses = []
    for _ in range(epochs):
        epoch_loss = 0
        for x, y in zip(X, Y):
            y_hat = params[0] * x + params[1]
            loss = (y - y_hat) ** 2
            epoch_loss += loss
            params[0] = params[0] - lr * (y_hat - y) * x
            params[1] = params[1] - lr * (y_hat - y)
        losses.append(epoch_loss)
    return params, losses
```

```python
params = torch.randn(2)
print(params.shape)
params, losses = sgd(params, 0.0001, X, y, 100)

plt.figure(figsize=(10, 5))
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.plot(np.log(losses))
```

```python
plt.figure(figsize=(10, 5))
plt.title('Dataset and Predicted Linear Regression')
plt.plot(X, y, 'o', label='data')
plt.plot([X[0], X[-1]], [X[0] * params[0] + params[1], X[-1] * params[0] + params[1]], 'r-', label='model')
plt.legend()
```
