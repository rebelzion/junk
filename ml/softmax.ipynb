```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
logits = torch.rand(10)
```

```python
plt.figure(figsize=(10, 5))
plt.title("Logits")
plt.plot(np.arange(logits.shape[0]), logits.detach().numpy(), '-')
```

```python
plt.figure(figsize=(10, 5))
plt.title('Gumbel Softmax')
for temperature in [0.001, 0.01, 0.1, 10, 100]:
    prob = torch.softmax(logits / temperature, dim=0)
    plt.plot(np.arange(prob.shape[0]), prob.detach().numpy(), '-', label=f"temperature = {temperature}")

plt.legend()
plt.plot()
```
